```python
weights_path = "C:\\Users\\wysy\\Downloads\\ёлка\\runs\\detect\\train6\\weights\\best.pt" # путь обученной модели
model = YOLO(weights_path) # использование веса
```

```python
folder_path = 'C:\\Users\\wysy\\OneDrive\\Рабочий стол\\втск\\Бутылки\\Разметка\\Пакеты для разметки\\Пак №1' # путь папки с фотографиями
```

```python
import os
images_list = os.listdir(folder_path) # создание списка путей фотографий
```

```python
import cv2
def visualize_pred(save_dir='C:\\Users\\wysy\\Downloads\\test_yolki'):
        for image in images_list:
                image_path = f'{folder_path}\\{image}'
                results = model(image_path)
                for result in results:
                        img = cv2.imread(image_path)
                        xyxys = result.boxes.xyxy.tolist()
                        classes = result.boxes.cls.tolist()
                        for num in range(len(xyxys)):
                                x = int(xyxys[num][0])
                                y = int(xyxys[num][1])
                                w = int(xyxys[num][2]) - int(xyxys[num][0])
                                h = int(xyxys[num][3]) - int(xyxys[num][1])
                                cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
                                cv2.putText(img, str(int(classes[num])), (x + w // 2,y + h // 2), cv2.FONT_HERSHEY_PLAIN, 1, (255,255,255), 2, 4)
                        cv2.imwrite(f'{save_dir}\\{image}', img)
```

```python
visualize_pred()
```
